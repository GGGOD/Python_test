```python
def pageDetail(detailurl,title,price):
    import requests
    from bs4 import BeautifulSoup
    res = requests.get(detailurl)
    soup = BeautifulSoup(res.text,'html.parser')
    html = soup.find('table',id="StructuredDataTable")
    data_list = []
    if  not html is None :
        aa = html.select('td')
        brand = aa[0].text
        series = aa[1].text
        screen_size = aa[2].text
        resolution = aa[3].text
        warranty = aa[4].text
        origin = aa[5].text
        item_dimension = aa[6].text
        refresh_rate = aa[11].text
        data_list = (brand,series,screen_size,resolution,warranty,origin,item_dimension,refresh_rate)
        return {'title': title,'price' : price,'brand' : brand,'series' : series,'screen_size': screen_size,'resolution': resolution,\
                'warranty': warranty,'origin': origin,'item_dimension' : item_dimension,'refresh_rate' : refresh_rate,'link':detailurl
                }
    else: return {'title': title,'price': price,'brand' : 'NA','series' : 'NA','screen_size': 'NA','resolution': 'NA',\
                'warranty': 'NA','origin': 'NA','item_dimension' : 'NA','refresh_rate' : 'NA','link':detailurl  }
```
