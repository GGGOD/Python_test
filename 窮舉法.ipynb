```python
for 牛頭人 in range(1,60):
    for 槌 in range(1,60):
        for 頭 in range(1,60):
            if 牛頭人 * 3+ 槌 * 3 == 30 and 牛頭人 + 槌+頭*2 == 20 and 頭 + 槌*4 == 9:
                print('牛頭人:' ,牛頭人,'槌:' , 槌, '頭:',頭)
                print('答案: ' ,頭+槌*牛頭人)
```
